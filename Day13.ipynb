```python
def solve_claw_machines(input_str):
    # Parse the input from the multiline string
    lines = [l for l in input_str.strip().split('\n') if l.strip() != '']
    # Each machine is described by 3 lines
    machines = [lines[i:i+3] for i in range(0, len(lines), 3)]

    total_cost = 0
    win_count = 0

    for m in machines:
        X_A, Y_A, X_B, Y_B, X_target, Y_target = parse_machine_block(m)
        cost = minimal_cost_to_win(X_A, Y_A, X_B, Y_B, X_target, Y_target)
        if cost is not None:
            win_count += 1
            total_cost += cost

    return win_count, total_cost

def parse_machine_block(lines):
    # Regex patterns
    pattern_button = r"Button [AB]: X\+(\d+), Y\+(\d+)"
    pattern_prize = r"Prize: X=(\d+), Y=(\d+)"

    A_line = lines[0].strip()
    B_line = lines[1].strip()
    P_line = lines[2].strip()

    A_match = re.match(pattern_button, A_line)
    B_match = re.match(pattern_button, B_line)
    P_match = re.match(pattern_prize, P_line)

    if not A_match or not B_match or not P_match:
        raise ValueError("Input format error.")

    X_A, Y_A = int(A_match.group(1)), int(A_match.group(2))
    X_B, Y_B = int(B_match.group(1)), int(B_match.group(2))
    X_target, Y_target = int(P_match.group(1)), int(P_match.group(2))

    return X_A, Y_A, X_B, Y_B, X_target, Y_target

def minimal_cost_to_win(X_A, Y_A, X_B, Y_B, X_target, Y_target):
    # Brute force all combinations of a,b in [0..100]
    best_cost = None
    for a in range(101):
        for b in range(101):
            if (X_A*a + X_B*b == X_target) and (Y_A*a + Y_B*b == Y_target):
                cost = 3*a + b
                if best_cost is None or cost < best_cost:
                    best_cost = cost
    return best_cost
```

```python
def solve_claw_machines_part_two(input_str):
    # Add the large offset
    OFFSET = 10000000000000
    
    lines = [l for l in input_str.strip().split('\n') if l.strip() != '']
    # Each machine: 3 lines
    machines = [lines[i:i+3] for i in range(0, len(lines), 3)]

    total_cost = 0
    win_count = 0

    for m in machines:
        X_A, Y_A, X_B, Y_B, X_target, Y_target = parse_machine_block(m)
        # Add the offset to the prize coordinates
        X_target += OFFSET
        Y_target += OFFSET
        cost = minimal_cost_to_win_large(X_A, Y_A, X_B, Y_B, X_target, Y_target)
        if cost is not None:
            win_count += 1
            total_cost += cost

    return win_count, total_cost

def minimal_cost_to_win_large(X_A, Y_A, X_B, Y_B, X_target, Y_target):
    # Solve the system:
    # X_A*a + X_B*b = X_target
    # Y_A*a + Y_B*b = Y_target

    # Determinant of the matrix
    D = X_A * Y_B - X_B * Y_A
    D_a = X_target * Y_B - X_B * Y_target
    D_b = X_A * Y_target - Y_A * X_target

    # If D == 0, either no solution or infinite solutions
    # For this puzzle, infinite solutions are unlikely. Check if no solution:
    if D == 0:
        # If any of D_a or D_b is not zero, no solution
        if D_a != 0 or D_b != 0:
            return None
        # If D=0 and also D_a=0 and D_b=0, infinite solutions:
        # This would mean any a,b that satisfies one eq satisfies another.
        # We must find non-negative integer solutions that minimize cost = 3*a+b.
        # Without additional constraints, let's assume no infinite solution scenario 
        # given puzzle context. If needed, we could handle it by scanning along one variable.
        return None

    # If D != 0, we have a unique solution:
    # a = D_a / D, b = D_b / D
    # Check for integrality:
    if D_a % D != 0 or D_b % D != 0:
        return None  # Not integer solution

    a = D_a // D
    b = D_b // D

    # We only consider non-negative solutions (can't press a button negative times)
    if a < 0 or b < 0:
        return None

    # Cost = 3*a + b
    cost = 3*a + b
    return cost
```
