```python
def count_xmas(grid):
    rows = len(grid)
    cols = len(grid[0])
    word = "XMAS"
    word_len = len(word)
    directions = [
        (0, 1),    # Right
        (0, -1),   # Left
        (1, 0),    # Down
        (-1, 0),   # Up
        (1, 1),    # Down-Right
        (1, -1),   # Down-Left
        (-1, 1),   # Up-Right
        (-1, -1)   # Up-Left
    ]
    
    count = 0
    
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols
    
    for r in range(rows):
        for c in range(cols):
            for dr, dc in directions:
                match = True
                for k in range(word_len):
                    nr, nc = r + k * dr, c + k * dc
                    if not is_valid(nr, nc) or grid[nr][nc] != word[k]:
                        match = False
                        break
                if match:
                    count += 1
                    
    return count
```

```python
def count_xmas_patterns(input_string):
    # Convert input string to a grid
    grid = [list(line) for line in input_string.splitlines()]
    rows = len(grid)
    cols = len(grid[0])
    count = 0

    # Function to check a single 3x3 region for valid X-MAS patterns
    def is_valid_xmas(r, c):
        # Ensure we stay within bounds of the grid
        if r + 2 >= rows or c + 2 >= cols:
            return False

        # Extract relevant characters for the "X" shape
        top_left = grid[r][c]
        top_right = grid[r][c + 2]
        middle = grid[r + 1][c + 1]
        bottom_left = grid[r + 2][c]
        bottom_right = grid[r + 2][c + 2]

        # Check all four valid patterns
        if (top_left, middle, bottom_right) == ("M", "A", "S") and \
           (top_right, middle, bottom_left) == ("M", "A", "S"):  # MAS-MAS
            return True
        if (top_left, middle, bottom_right) == ("S", "A", "M") and \
           (top_right, middle, bottom_left) == ("S", "A", "M"):  # SAM-SAM
            return True
        if (top_left, middle, bottom_right) == ("M", "A", "S") and \
           (top_right, middle, bottom_left) == ("S", "A", "M"):  # MAS-SAM
            return True
        if (top_left, middle, bottom_right) == ("S", "A", "M") and \
           (top_right, middle, bottom_left) == ("M", "A", "S"):  # SAM-MAS
            return True

        return False

    # Traverse the grid for all possible 3x3 regions
    for r in range(rows - 2):
        for c in range(cols - 2):
            if is_valid_xmas(r, c):
                count += 1  # Increment count for every valid pattern found

    return count


# Example input as a multiline string
example_input = input_string.strip()

# Count X-MAS patterns in the example input
result = count_xmas_patterns(example_input)
print(f"Total occurrences of X-MAS patterns: {result}")
```
